```python
import numpy as np
import matplotlib.pyplot as mp
x=[2.0,9.4,3.32,0.88,-2.23,1.11,0.57,-2.25,-3.31,6.45]
y=[1.22,0.34,-0.08,2.25,4.41,3.09,-6.66,-9.77,0.001,2.25]
predict_y=[]
mean_x = np.mean(x)
mean_y = np.mean(y)

var_x = np.var(x)

covar_x_y = np.cov(x,y,bias=True)[0][1]

m = covar_x_y/var_x

c = mean_y-(m*mean_x)

print(mean_x)
print(mean_y)
print(var_x)
print(covar_x_y)
print(m)
print(c)
```

```python
for i in range(0,10):
    new_y=(m*x[i])+c
    predict_y.append(new_y)
print(predict_y)
mp.plot(y,predict_y)
mp.show()
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets) ** 2).mean())

rmse_val = rmse(np.array(y), np.array(predict_y))
print("Root Mean Square error is: " + str(rmse_val))
```
