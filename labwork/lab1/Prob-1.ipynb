```python
import tensorflow as tf
a = tf.constant(5)
b = tf.constant(4)
with tf.name_scope("a_plus_b_squared"):
    with tf.name_scope("Scope1"):
        c=tf.multiply(a,a,name="25_squared")
        d=tf.multiply(b,b,name="16_squared")
with tf.name_scope("Scope2"):
    e=tf.add(c,d,name="Addition_a_squared_and_b_squared_performed_Ans_41")
    f=tf.multiply(a,b,name="Multiplication_of_a_and_b_performed_Ans_20")
with tf.name_scope("Scope3"):
    g=tf.multiply(2,f,name="Multiplication_2ab_performed_Ans_40")
i=tf.add(e,g,name="Total_ans_81")

with tf.Session() as ssn:
    writer = tf.summary.FileWriter("/tmp/tboard/output5",ssn.graph)
    print(ssn.run(i))
    writer.close()
```
