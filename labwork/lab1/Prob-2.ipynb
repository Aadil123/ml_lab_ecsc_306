```python
import tensorflow as tf
a = tf.constant(6)
b = tf.constant(2)
c = tf.constant(3)
with tf.name_scope("a_plus_b_plus_c_squared"):
    with tf.name_scope("Scope1"):
        d=tf.multiply(a,a,name="36_squared")
        e=tf.multiply(b,b,name="4_squared")
        f=tf.multiply(c,c,name="9_squared")
with tf.name_scope("Scope2"):
    g=tf.add(d,e,name="Addition_a_squared_and_b_squared_performed_Ans_40")
    h=tf.add(g,f,name="Addition_of_a_squared_and_b_squared_and_c_squared_performed_Ans_49")
with tf.name_scope("Scope3"):
    i=tf.multiply(a,b,name="Multiplication_ab_performed_Ans_12")
    j=tf.multiply(2,i,name="Multiplication_2ab_performed_Ans_24")
with tf.name_scope("Scope4"):
    k=tf.multiply(b,c,name="Multiplication_bc_performed_Ans_6")
    l=tf.multiply(2,k,name="Multiplication_2bc_performed_Ans_12")
with tf.name_scope("Scope5"):
    m=tf.multiply(c,a,name="Multiplication_ca_performed_Ans_18")
    n=tf.multiply(2,m,name="Multiplication_2ca_performed_Ans_36")
o=tf.add(j,l,name="2ab_plus_2bc_Ans_36")
p=tf.add(n,o,name="2ab_plus_2bc_plus_2ca_Ans_72")
q=tf.add(h,p,name="Total_ans_121")

with tf.Session() as ssn:
    writer = tf.summary.FileWriter("/tmp/tboard/output240",ssn.graph)
    print(ssn.run(q))
    writer.close()
```
